### Dataset Description

Neurologisk Klinik has compiled a synthetic and anonymized dataset [1] with detailed health information for synthetic 2,149 patients, each identified by a unique ID (4751 and 6900). The dataset contains various types of data. Patient Information includes a unique PatientID (integer). Demographic Details cover Age (integer), Gender (binary), Ethnicity, and EducationLevel (categorical). Lifestyle Factors include BMI (float), Smoking, AlcoholConsumption, PhysicalActivity, DietQuality, and SleepQuality (integers or binary). Medical History variables are binary (e.g., FamilyHistoryAlzheimers, Hypertension). Clinical Measurements include blood pressure and cholesterol levels (integers). Cognitive and Functional Assessments cover MMSE, FunctionalAssessment, ADL (integers), and several binary indicators. Symptoms are binary (e.g., Confusion, Forgetfulness). Diagnosis Information is binary for Alzheimer’s status. DoctorInCharge: This column contains confidential information about the doctor in charge, with "XXXConfid" as the value for all patients.

The original dataset can be retrieved from: https://www.kaggle.com/datasets/rabieelkharoua/alzheimers-disease-dataset [1]

#### Correlation among Categorical Variables using Chi square test

````python
import pandas as pd
from scipy.stats import chi2_contingency

# Function to perform Chi-Square test and return results in a DataFrame
def chi_square_test(df, cat_vars):
    # Create an empty dictionary to store results
    results = {
        'Variable 1': [],
        'Variable 2': [],
        'Chi-Square Stat': [],
        'Degrees of Freedom': [],
        'P-Value': []
    }
    
    # Loop through all pairs of categorical variables
    for i in range(len(cat_vars)):
        for j in range(i+1, len(cat_vars)):
            var1 = cat_vars[i]
            var2 = cat_vars[j]
            
            # Create a contingency table
            contingency_table = pd.crosstab(df[var1], df[var2])
            
            # Perform the Chi-Square test
            chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)
            
            # Store the results
            results['Variable 1'].append(var1)
            results['Variable 2'].append(var2)
            results['Chi-Square Stat'].append(round(chi2_stat, 4))  # Round for readability
            results['Degrees of Freedom'].append(dof)
            results['P-Value'].append(round(p_value, 4))  # Round for readability
    
    # Convert to a DataFrame for better visualization
    results_df = pd.DataFrame(results)
    
    return results_df

# List of categorical variables to analyze
categorical_vars = ['Gender', 'EducationLevel', 'Ethnicity', "ImpairmentLevel", "DependencyLevel", "WeightStatus", "Smoking"]

# Run Chi-Square test
chi_square_results = chi_square_test(df_dataviz, categorical_vars)

# Filter for p-value < 0.05
filtered_results = chi_square_results[chi_square_results['P-Value'] < 0.05]

# Display the filtered table of results
print(filtered_results)
````

#### Notes for Chi Square Results
Chi-Square Stat: This is the chi-square statistic, which measures the discrepancy between the observed and expected frequency counts in the contingency table.

Degrees of Freedom: This is the number of independent values or quantities that can vary in the test.

P-Value: This is the probability of obtaining the observed chi-square statistic, or one more extreme, assuming the null hypothesis is true (no association between the two categorical variables).

````python
# Trying 2 until 5 clusters, and compare which one is better

# Obs: Although I've sampled the population as shown above, this chunk still takes some minutes to run (2-3 min in my testing sections)

clusters_to_evaluate = np.arange(2,6) # Generates a sequence: I've choosen run 2 to 5 clusters
print(f"Testing num_clusters: {clusters_to_evaluate}")

# Store the results in a dataframe, initialize empty array
results = pd.DataFrame({"k": [],
                        "silhouette_score": []})

# Iterate
for k in clusters_to_evaluate:
    # Note that the parameter `n_clusters` is the variable `k`
    kmeans = KMeans(n_clusters = k, random_state=RANDOM_SEED, init='random', n_init='auto', max_iter=100)
    # Generate cluster labels for the data
    predicted_clusters = kmeans.fit_predict(df_normalized)
    # Calculate silhouette score
    sil_score = metrics.silhouette_score(df_normalized, predicted_clusters)

    # Add the results to the pandas dataframe
    this_result = pd.DataFrame({"k": [k],
                                "silhouette_score": [sil_score]})

    results = pd.concat([results, this_result], ignore_index=True)

results["k"] = results["k"].astype(np.int16)

# Print results
print("The silhouette score for each number of clusters is presented below:")
print(results)
````

### Evaluation of Clustering Models

#### Summary of Metrics
Silhouette Score: Measures how similar an instance is to its own cluster vs. other clusters (ranges from -1 to +1).

Davies-Bouldin Index: Measures the average similarity between clusters (lower is better).

Calinski-Harabasz Index: Measures the separation and cohesion of clusters (higher is better).

````python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
import pandas as pd

# Store clustering evaluation results in a dictionary
evaluation_metrics = {
    'Clustering Algorithm': [],
    'Silhouette Score': [],
    'Davies-Bouldin Index': [],
    'Calinski-Harabasz Index': []
}

# Iterate through the clustering models
for name, model in clustering_models.items():
    if name == 'Gaussian Mixture':
        clusters = model.fit_predict(df_scaled)
    else:
        clusters = model.fit_predict(df_scaled)
    
    # Calculate each metric
    silhouette_avg = silhouette_score(df_scaled, clusters)
    db_index = davies_bouldin_score(df_scaled, clusters)
    ch_index = calinski_harabasz_score(df_scaled, clusters)
    
    # Store the results for comparison
    evaluation_metrics['Clustering Algorithm'].append(name)
    evaluation_metrics['Silhouette Score'].append(silhouette_avg)
    evaluation_metrics['Davies-Bouldin Index'].append(db_index)
    evaluation_metrics['Calinski-Harabasz Index'].append(ch_index)

# Convert to a DataFrame for better manipulation and visualization
evaluation_df = pd.DataFrame(evaluation_metrics)

# Visualization: Grouped bar plots for each metric
metrics = ['Silhouette Score', 'Davies-Bouldin Index', 'Calinski-Harabasz Index']

# Create subplots only for 3 metrics
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('Clustering Evaluation Metrics Comparison', fontsize=16)

# Silhouette Score
sns.barplot(x='Clustering Algorithm', y='Silhouette Score', data=evaluation_df, ax=axes[0])
axes[0].set_title('Silhouette Score')
axes[0].set_ylim(0, 1)  # Since silhouette score ranges from -1 to 1, we restrict to 0 to 1

# Davies-Bouldin Index
sns.barplot(x='Clustering Algorithm', y='Davies-Bouldin Index', data=evaluation_df, ax=axes[1])
axes[1].set_title('Davies-Bouldin Index')
axes[1].set_ylim(0, evaluation_df['Davies-Bouldin Index'].max() + 1)

# Calinski-Harabasz Index
sns.barplot(x='Clustering Algorithm', y='Calinski-Harabasz Index', data=evaluation_df, ax=axes[2])
axes[2].set_title('Calinski-Harabasz Index')
axes[2].set_ylim(0, evaluation_df['Calinski-Harabasz Index'].max() + 500)

# Adjust layout to avoid overlap and add space for text
plt.tight_layout()
plt.subplots_adjust(top=0.85, bottom=0.2)  # Move the plots down to make space for the text

# Add explanatory text below the plots
text = (
    "Silhouette Score: Measures how similar an object is to its own cluster compared to other clusters. A higher score indicates better-defined clusters.\n"
    "Davies-Bouldin Index: Measures the average similarity ratio of each cluster to the cluster that is most similar to it. A lower value indicates better clustering.\n"
    "Calinski-Harabasz Index: Measures the ratio of the sum of between-cluster dispersion to within-cluster dispersion. A higher value indicates better clustering."
)

# Place the text at the bottom of the figure (use y=0.02 for more spacing)
plt.figtext(0.5, 0.02, text, wrap=True, horizontalalignment='center', fontsize=10)

# Show the plot
plt.show()
````
