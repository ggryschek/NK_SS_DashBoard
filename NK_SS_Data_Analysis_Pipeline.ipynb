## Dataset Description

Neurologisk Klinik has compiled a synthetic and anonymized dataset [1] with detailed health information for synthetic 2,149 patients, each identified by a unique ID (4751 and 6900). The dataset contains various types of data. Patient Information includes a unique PatientID (integer). Demographic Details cover Age (integer), Gender (binary), Ethnicity, and EducationLevel (categorical). Lifestyle Factors include BMI (float), Smoking, AlcoholConsumption, PhysicalActivity, DietQuality, and SleepQuality (integers or binary). Medical History variables are binary (e.g., FamilyHistoryAlzheimers, Hypertension). Clinical Measurements include blood pressure and cholesterol levels (integers). Cognitive and Functional Assessments cover MMSE, FunctionalAssessment, ADL (integers), and several binary indicators. Symptoms are binary (e.g., Confusion, Forgetfulness). Diagnosis Information is binary for Alzheimer’s status. DoctorInCharge: This column contains confidential information about the doctor in charge, with "XXXConfid" as the value for all patients.

The original dataset can be retrieved from: https://www.kaggle.com/datasets/rabieelkharoua/alzheimers-disease-dataset

### Data Cleaning and Wrangling

````python
# PatientID is a category rather than a Integer and will be correctly assigned

df['PatientID'] = df['PatientID'].astype("category")
````

````python
# Attribute a Doctor to the patients (randomly at this point, as in the dataset it is confidential) - it will be used for data vizualization purposes

# Set a random seed for reproducibility
np.random.seed(42)

# Include the random Doctor In Charge variable

df['DoctorInCharge'] = np.random.choice(['Med A', 'Med B', 'Med C'], size=len(df))

# Convert the column to categorical
df['DoctorInCharge'] = pd.Categorical(df['DoctorInCharge'], categories=['Med A', 'Med B', 'Med C'])

# Check the distribution
print(df['DoctorInCharge'].value_counts())
````

````python
# Insert categories related to scores: Impairment (MMSE)
# Step 1: Create a new categorical variable based on the MMSE values: (≥ 27	None; 21-26	Mild; 11-20	Moderate; ≤ 10	Severe)

df['ImpairmentLevel'] = pd.cut(
    df['MMSE'], 
    bins=[-float('inf'), 10, 20, 26, float('inf')],  # Define the cut-off points
    labels=['Severe', 'Moderate', 'Mild', 'None'])

# Step 2: Convert to an ordinal categorical variable
df['ImpairmentLevel'] = pd.Categorical(
    df['ImpairmentLevel'], 
    categories=['Severe', 'Moderate', 'Mild', 'None'], ordered=True)
````

````python
# Insert categories related to scores: Funcionality (ADL - Barthel Index)
# Step 1: Create a new categorical variable based on the ADL values: Barthel Index (8.0–10.0: Independent; 6.0–7.9: Minimally dependent;
# 4.0–5.9: Partially dependent; 2.0–3.9: Very dependent; <2.0: Totally dependent)

df['DependencyLevel'] = pd.cut(
    df['ADL'], 
    bins=[-float('inf'), 1.9, 3.9, 5.9, 7.9, 10],  # Define the cut-off points
    labels=['Totally dependent', 'Very dependent', 'Partially dependent', 'Minimally dependent', 'Independent'])

# Step 2: Convert to an ordinal categorical variable
df['DependencyLevel'] = pd.Categorical(
    df['DependencyLevel'], 
    categories=['Totally dependent', 'Very dependent', 'Partially dependent', 'Minimally dependent', 'Independent'], ordered=True)
````

````python
# Round all float variables to 2 decimals, for better stardatization (all the real observation would have no more than 2 decimals)

# Step 1: List of columns to be rounded
columns_decimal = ['BMI', 'ADL', 'CholesterolTotal','CholesterolLDL','CholesterolHDL','CholesterolTriglycerides']
columns_integer = ['AlcoholConsumption', 'PhysicalActivity', 'DietQuality', 'SleepQuality', 'MMSE', 'FunctionalAssessment']

# Step 2: Apply rounding to only the selected columns
df[columns_decimal] = df[columns_decimal].round(2)

df[columns_integer] = df[columns_integer].round(0).astype(int)
````

````python
# We will keep the dataset for ML analysis and create a copy for data wrangling to support dataviz
df_dataviz = df
````

````python
#Many features were converted as integer datatype but most of them are categorical: binary boolean or ordinal
#Explicitly convert the binary categorical variables: first replacing the 'yes' and 'no' values to True and False and then using bool type
#Converting binary object datatypes into binary boolean

# List of columns to convert
columns_to_convert = ['Smoking','FamilyHistoryAlzheimers', 'CardiovascularDisease','Depression','Diabetes','HeadInjury', 'Hypertension', 'MemoryComplaints',
                      'BehavioralProblems', 'Confusion', 'Disorientation', 'PersonalityChanges', 'DifficultyCompletingTasks', 'Forgetfulness', 'Diagnosis']

# Define the conversion mapping
conversion_mapping = {1: True, 2: False}

# Loop over the columns and perform conversion
for column in columns_to_convert:
    df_dataviz[column] = df_dataviz[column].replace(conversion_mapping).astype(bool)
````

````python
# Converting the nominal category Gender
df_dataviz['Gender'] = df_dataviz['Gender'].astype("category")

#Transform the binary feature `gender` into numerical values
df_dataviz['Gender'] = df_dataviz['Gender'].map({
                0:"Female",
                1:"Male"})
````

````python
#Converting "EducationalLevel" feature to ordinal category
educ_level = ["None", "High School", "Bachelor's", "Higher"]

df_dataviz['EducationLevel'] = df_dataviz['EducationLevel'].map({0: "None", 1: "High School", 2: "Bachelor's", 3: "Higher"})

df_dataviz["EducationLevel"] = pd.Categorical(df_dataviz["EducationLevel"], categories=educ_level, ordered=True)
````

````python
#Converting "Ethnicity" feature to nominal category
ethinicity = ["Caucasian", "Black", "Asian", "Other"]

df_dataviz['Ethnicity'] = df_dataviz['Ethnicity'].map({0: "Caucasian", 1: "Black", 2: "Asian", 3: "Other"})

df_dataviz["Ethnicity"] = pd.Categorical(df_dataviz["Ethnicity"], categories=ethinicity, ordered=False)
````

````python
# List of boolean variables to analyze
boolean_variables = ['MemoryComplaints', 'BehavioralProblems', 'Confusion', 
                     'Disorientation', 'PersonalityChanges', 
                     'DifficultyCompletingTasks', 'Forgetfulness']

# Calculate the total number of cases (rows) in the dataset
total_cases = len(dataviz_ADcount)

# Calculate the number of True values and proportions for each boolean variable
true_counts = []
proportions = []

for var in boolean_variables:
    true_count = dataviz_ADcount[var].sum()
    proportion = f"{round((true_count / total_cases) * 100, 2)}%"
    true_counts.append(true_count)
    proportions.append(proportion)

# Create a DataFrame to store the results
summary_df = pd.DataFrame({
    'Condition': ['Memory Complaints', 'Behavioral Problems', 'Confusion', 
                  'Disorientation', 'Personality Changes', 
                  'Difficulty Completing Tasks', 'Forgetfulness'],
    'Complaint Count': true_counts,
    'Proportion': proportions
})

# Display the proportion table
print(summary_df)

# Create a bar chart to compare the counts of True values
fig = px.bar(summary_df, x='Condition', y='Complaint Count', 
             title='Comparison of Complaint Count for Various Conditions',
             labels={'Condition': 'Condition', 'Complaint Count': 'Cases'},
             text='Complaint Count',
             color='Condition')

# Show the figure
fig.show()
````

````python
# Creating a dynamic distribution chart with all filters for end users
class DashApp:
    def __init__(self, dataframe):
        self.df = dataframe
        self.app = dash.Dash('DashApp', suppress_callback_exceptions=True)
        self.filterable_vars = ['Gender', 'DoctorInCharge', 'EducationLevel', 'Ethnicity']
        self.available_columns = [col for col in self.df.columns if col != 'PatientID']  # Exclude PatientID

    def generate_table(self, filtered_df, selected_columns):
        table_columns = ['PatientID'] + self.filterable_vars + [col for col in selected_columns if col not in self.filterable_vars]
        return html.Table([
            html.Thead(html.Tr([html.Th(col, 
            style={'color': 'black', 'backgroundColor': '#0074D9', 'fontFamily': 'Arial', 'fontSize': '16px', 'textAlign': 'center'}) for col in table_columns])),
            html.Tbody([html.Tr([html.Td(str(filtered_df.iloc[i][col]), 
            style={'color': 'black', 'backgroundColor': '#f9f9f9', 'fontFamily': 'Arial', 'fontSize': '14px', 'textAlign': 'center'}) for col in table_columns]) for i in range(min(len(filtered_df), 10))])
        ])

    def generate_histogram(self, filtered_df):
        return dcc.Graph(
            figure={
                'data': [
                    {
                        'x': filtered_df['Age'],
                        'type': 'histogram',
                        'marker': {'color': '#0074D9'},
                        'name': 'Age Distribution'
                    }
                ],
                'layout': {
                    'title': 'Age Distribution',
                    'xaxis': {'title': 'Age'},
                    'yaxis': {'title': 'Count'},
                    'paper_bgcolor': '#f9f9f9',
                    'plot_bgcolor': '#f9f9f9'
                }
            }
        )

    def layout(self):
        # Generate filter dropdowns based on the specified filterable variables
        filter_dropdowns = [
            html.Div([
                html.Label(f"Filter by {var}"),
                dcc.Dropdown(
                    id=f'{var}-filter',
                    options=[{'label': val, 'value': val} for val in self.df[var].unique()],
                    value=None,
                    placeholder=f'Select {var} value',
                    multi=True,
                    style={'backgroundColor': '#f0f0f0', 'color': 'black'}
                )
            ]) for var in self.filterable_vars
        ]

        self.app.layout = html.Div([
            html.H1("Patient Data by Age Distribution"),

            # Filter dropdowns in the specified order
            *filter_dropdowns,

            # Range slider for Age filter
            dcc.RangeSlider(
                id='age-slider',
                min=self.df['Age'].min(),
                max=self.df['Age'].max(),
                marks={i: str(i) for i in range(int(self.df['Age'].min()), int(self.df['Age'].max()) + 1, 5)},
                value=[self.df['Age'].min(), self.df['Age'].max()],
                step=1
            ),

            # Dropdown for PatientID selection
            dcc.Dropdown(
                id='patient-filter',
                options=[{'label': str(pid), 'value': pid} for pid in self.df['PatientID'].unique()],
                value=None,
                placeholder='Select a Patient ID',
                multi=False,
                style={'backgroundColor': '#f0f0f0', 'color': 'black'}
            ),

            # Dropdown for column selection (multi-select), allowing user to include/exclude columns
            dcc.Dropdown(
                id='column-filter',
                options=[{'label': col, 'value': col} for col in self.available_columns],
                value=self.available_columns,  # By default, show all available columns
                multi=True,
                style={'backgroundColor': '#f0f0f0', 'color': 'black'}
            ),

            # Histogram for Age distribution
            html.Div(id='histogram-container'),

            # Table to display the filtered data
            html.Div(id='table-container')
        ])

    def setup_callbacks(self):
        @self.app.callback(
            [Output('table-container', 'children'),
             Output('histogram-container', 'children')],
            [Input('column-filter', 'value'), 
             Input('patient-filter', 'value'),
             Input('age-slider', 'value')] +
            [Input(f'{var}-filter', 'value') for var in self.filterable_vars]  # Add inputs for dynamic filters
        )
        def update_output(selected_columns, selected_patient, selected_age_range, *filter_values):
            # Filter by selected patient
            filtered_df = self.df.copy()
            if selected_patient:
                filtered_df = filtered_df[filtered_df['PatientID'] == selected_patient]
            
            # Filter by selected age range
            filtered_df = filtered_df[
                (filtered_df['Age'] >= selected_age_range[0]) & 
                (filtered_df['Age'] <= selected_age_range[1])
            ]
            
            # Apply filters based on selected filterable variables
            for i, var in enumerate(self.filterable_vars):
                if filter_values[i]:
                    filtered_df = filtered_df[filtered_df[var].isin(filter_values[i])]

            # Generate table and histogram with the filtered data
            table = self.generate_table(filtered_df, selected_columns)
            histogram = self.generate_histogram(filtered_df)
            
            return table, histogram

    def run(self):
        self.layout()
        self.setup_callbacks()
        self.app.run_server(debug=True)
````

````python
# Create Vizualization for Medical Condition and Symptoms Distribution
class BarChartApp:
    def __init__(self, df):
        self.df = df
        self.app = dash.Dash('BarChartApp', suppress_callback_exceptions=True)
        self.filterable_vars = ['DoctorInCharge', 'Gender','EducationLevel', 'Ethnicity']  # Filters for user control
        excluded_columns = ['Diagnosis']
        self.boolean_vars = [col for col in df.columns if df[col].dtype == 'bool' and col not in excluded_columns]  # Boolean columns in the dataset
        self.setup_layout()
        self.setup_callbacks()

    def setup_layout(self):
        # Layout for filters, stacked bar chart, and patient table
        self.app.layout = html.Div([
            html.H1("Distribution of Medical Conditions, Behaviour and Symptoms"),

            # Dropdown for selecting the boolean condition to visualize
            dcc.Dropdown(
                id='boolean-condition',
                options=[{'label': var, 'value': var} for var in self.boolean_vars],
                value=self.boolean_vars[0],  # Default to the first boolean variable
                placeholder="Select a condition",
                multi=False,
                style={'backgroundColor': '#f0f0f0', 'color': 'black'}
            ),

            # Range slider for Age filter
            dcc.RangeSlider(
                id='age-slider',
                min=self.df['Age'].min(),
                max=self.df['Age'].max(),
                marks={i: str(i) for i in range(int(self.df['Age'].min()), int(self.df['Age'].max()) + 1, 5)},
                value=[self.df['Age'].min(), self.df['Age'].max()],
                step=1
            ),

            # Filter dropdowns for additional demographic data
            *[
                html.Div([
                    html.Label(f"Filter by {var}"),
                    dcc.Dropdown(
                        id=f'{var}-filter',
                        options=[{'label': val, 'value': val} for val in self.df[var].unique()],
                        value=None,
                        placeholder=f'Select {var}',
                        multi=True,
                        style={'backgroundColor': '#f0f0f0', 'color': 'black'}
                    )
                ]) for var in self.filterable_vars
            ],

            # Stacked bar chart display
            html.Div(id='stacked-bar-container'),

            # Table display for patients with True condition
            html.Div(id='true-condition-table-container')
        ])

    def setup_callbacks(self):
        @self.app.callback(
            [Output('stacked-bar-container', 'children'),
             Output('true-condition-table-container', 'children')],
            [Input('boolean-condition', 'value'),
             Input('age-slider', 'value')] +
            [Input(f'{var}-filter', 'value') for var in self.filterable_vars]
        )
        def update_bar_table(selected_condition, selected_age_range, *filter_values):
            # Start with full DataFrame
            filtered_df = self.df.copy()

            # Apply age range filter
            filtered_df = filtered_df[
                (filtered_df['Age'] >= selected_age_range[0]) &
                (filtered_df['Age'] <= selected_age_range[1])
            ]

            # Apply filters for Gender, DoctorInCharge, EducationLevel, and Ethnicity
            for i, var in enumerate(self.filterable_vars):
                if filter_values[i]:
                    filtered_df = filtered_df[filtered_df[var].isin(filter_values[i])]

            # Generate stacked bar chart for the selected boolean condition
            stacked_bar_chart = self.generate_bar_chart(filtered_df, selected_condition)

            # Generate table for patients with True condition
            true_condition_table = self.generate_true_condition_table(filtered_df, selected_condition)

            return stacked_bar_chart, true_condition_table

    def generate_bar_chart(self, filtered_df, selected_condition):
        # Create a stacked bar chart with counts of True/False for the selected boolean condition
        chart_data = filtered_df.groupby([selected_condition]).size().reset_index(name='count')

        fig = px.bar(
            chart_data,
            x=selected_condition,
            y='count',
            color=selected_condition,
            title=f"Distribution of {selected_condition}",
            barmode='stack',
            labels={selected_condition: 'Condition', 'count': 'Count'},
            color_discrete_map={True: '#0074D9', False: '#FF4136'}  # Customize colors for True/False
        )

        return dcc.Graph(figure=fig)

    def generate_true_condition_table(self, filtered_df, selected_condition):
        # Filter the dataframe for rows where the selected condition is True
        true_condition_df = filtered_df[filtered_df[selected_condition] == True]

        # Limit to first 10 rows for display
        true_condition_df = true_condition_df.head(10)

        # Generate a simple HTML table to display the patients with True condition
        return html.Table([
    # Table Header with styling
            html.Thead(html.Tr([html.Th(col, style={
                'backgroundColor': '#0074D9', 'color': 'white', 'padding': '10px', 'textAlign': 'center',
                'border': '1px solid black', 'fontWeight': 'bold', 'fontSize': '16px', 'fontFamily': 'Arial'
            }) for col in true_condition_df.columns])),

            # Table Body with styling
            html.Tbody([
                html.Tr([html.Td(str(true_condition_df.iloc[i][col]), style={
                    'backgroundColor': '#f9f9f9', 'color': 'black', 'padding': '8px', 'textAlign': 'center',
                    'border': '1px solid black', 'fontFamily': 'Arial', 'fontSize': '14px'
                }) for col in true_condition_df.columns])
                for i in range(len(true_condition_df))
            ])
        ])

    def run(self):
        self.app.run_server(debug=True)
````
