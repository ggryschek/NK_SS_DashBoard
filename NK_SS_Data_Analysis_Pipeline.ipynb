## Dataset Description

Neurologisk Klinik has compiled an anonymized dataset [1] with detailed health information for 2,149 patients, each identified by a unique ID (4751 and 6900). The dataset contains various types of data. Patient Information includes a unique PatientID (integer). Demographic Details cover Age (integer), Gender (binary), Ethnicity, and EducationLevel (categorical). Lifestyle Factors include BMI (float), Smoking, AlcoholConsumption, PhysicalActivity, DietQuality, and SleepQuality (integers or binary). Medical History variables are binary (e.g., FamilyHistoryAlzheimers, Hypertension). Clinical Measurements include blood pressure and cholesterol levels (integers). Cognitive and Functional Assessments cover MMSE, FunctionalAssessment, ADL (integers), and several binary indicators. Symptoms are binary (e.g., Confusion, Forgetfulness). Diagnosis Information is binary for Alzheimer’s status. DoctorInCharge: This column contains confidential information about the doctor in charge, with "XXXConfid" as the value for all patients.

##### Details - Patient Information
1. Patient ID

        1.1 PatientID: A unique identifier assigned to each patient (4751 to 6900).

2. Demographic Details

        2.1 Age: The age of the patients ranges from 60 to 90 years.
        2.2 Gender: Gender of the patients, where 0 represents Male and 1 represents Female.
        2.3 Ethnicity: The ethnicity of the patients, coded as follows:
            0: Caucasian
            1: African American
            2: Asian
            3: Other
        2.4 EducationLevel: The education level of the patients, coded as follows:
            0: None
            1: High School
            2: Bachelor's
            3: Higher
            
3. Lifestyle Factors

        3.1 BMI: Body Mass Index of the patients, ranging from 15 to 40.
        3.2 Smoking: Smoking status, where 0 indicates No and 1 indicates Yes.
        3.4 AlcoholConsumption: Weekly alcohol consumption in units, ranging from 0 to 20.
        3.5 PhysicalActivity: Weekly physical activity in hours, ranging from 0 to 10.
        3.6 DietQuality: Diet quality score, ranging from 0 to 10.
        3.7 SleepQuality: Sleep quality score, ranging from 4 to 10.

4. Medical History

        4.1 FamilyHistoryAlzheimers: Family history of Alzheimer's Disease, where 0 indicates No and 1 indicates Yes.
        4.2 CardiovascularDisease: Presence of cardiovascular disease, where 0 indicates No and 1 indicates Yes.
        4.3 Diabetes: Presence of diabetes, where 0 indicates No and 1 indicates Yes.
        4.4 Depression: Presence of depression, where 0 indicates No and 1 indicates Yes.
        4.5 HeadInjury: History of head injury, where 0 indicates No and 1 indicates Yes.
        4.6 Hypertension: Presence of hypertension, where 0 indicates No and 1 indicates Yes.

5. Clinical Measurements

        5.1 SystolicBP: Systolic blood pressure, ranging from 90 to 180 mmHg.
        5.2 DiastolicBP: Diastolic blood pressure, ranging from 60 to 120 mmHg.
        5.3 CholesterolTotal: Total cholesterol levels, ranging from 150 to 300 mg/dL.
        5.4 CholesterolLDL: Low-density lipoprotein cholesterol levels, ranging from 50 to 200 mg/dL.
        5.5 CholesterolHDL: High-density lipoprotein cholesterol levels, ranging from 20 to 100 mg/dL.
        5.6 CholesterolTriglycerides: Triglycerides levels, ranging from 50 to 400 mg/dL.

6. Cognitive and Functional Assessments

        6.1 MMSE: Mini-Mental State Examination score, ranging from 0 to 30. Lower scores indicate cognitive impairment.
        6.2 FunctionalAssessment: Functional assessment score, ranging from 0 to 10. Lower scores indicate greater impairment.
        6.3 MemoryComplaints: Presence of memory complaints, where 0 indicates No and 1 indicates Yes.
        6.4 BehavioralProblems: Presence of behavioral problems, where 0 indicates No and 1 indicates Yes.
        6.5 ADL: Activities of Daily Living score, ranging from 0 to 10.00. Lower scores indicate greater impairment.

7. Symptoms

        7.1 Confusion: Presence of confusion, where 0 indicates No and 1 indicates Yes.
        7.2 Disorientation: Presence of disorientation, where 0 indicates No and 1 indicates Yes.
        7.3 PersonalityChanges: Presence of personality changes, where 0 indicates No and 1 indicates Yes.
        7.4 DifficultyCompletingTasks: Presence of difficulty completing tasks, where 0 indicates No and 1 indicates Yes.
        7.5 Forgetfulness: Presence of forgetfulness, where 0 indicates No and 1 indicates Yes.

8. Diagnosis Information

        8.1 Diagnosis: Diagnosis status for Alzheimer's Disease, where 0 indicates No and 1 indicates Yes.

9. Confidential Information

        9.1 DoctorInCharge: This column contains confidential information about the doctor in charge, with "XXXConfid" as the value for all patients.

### Data Cleaning and Wrangling

````python
# PatientID is a category rather than a Integer and will be correctly assigned

df['PatientID'] = df['PatientID'].astype("category")
````

````python
# Attribute a Doctor to the patients (randomly at this point, as in the dataset it is confidential)

# Set a random seed for reproducibility
np.random.seed(42)

# Include the random Doctor In Charge variable

df['DoctorInCharge'] = np.random.choice(['Med A', 'Med B', 'Med C'], size=len(df))

# Convert the column to categorical
df['DoctorInCharge'] = pd.Categorical(df['DoctorInCharge'], categories=['Med A', 'Med B', 'Med C'])

# Check the distribution
print(df['DoctorInCharge'].value_counts())
````

````python
# Insert categories related to scores: Impairment (MMSE)
# Step 1: Create a new categorical variable based on the MMSE values: (≥ 27	None; 21-26	Mild; 11-20	Moderate; ≤ 10	Severe)

df['ImpairmentLevel'] = pd.cut(
    df['MMSE'], 
    bins=[-float('inf'), 10, 20, 26, float('inf')],  # Define the cut-off points
    labels=['Severe', 'Moderate', 'Mild', 'None'])

# Step 2: Convert to an ordinal categorical variable
df['ImpairmentLevel'] = pd.Categorical(
    df['ImpairmentLevel'], 
    categories=['Severe', 'Moderate', 'Mild', 'None'], ordered=True)
````

````python
# Insert categories related to scores: Funcionality (ADL - Barthel Index)
# Step 1: Create a new categorical variable based on the ADL values: Barthel Index (8.0–10.0: Independent; 6.0–7.9: Minimally dependent
# 4.0–5.9: Partially dependent; 20–39: Very dependent; <20: Totally dependent)

df['DependencyLevel'] = pd.cut(
    df['ADL'], 
    bins=[-float('inf'), 1.9, 3.9, 5.9, 7.9, 10],  # Define the cut-off points
    labels=['Totally dependent', 'Very dependent', 'Partially dependent', 'Minimally dependent', 'Independent'])

# Step 2: Convert to an ordinal categorical variable
df['DependencyLevel'] = pd.Categorical(
    df['DependencyLevel'], 
    categories=['Totally dependent', 'Very dependent', 'Partially dependent', 'Minimally dependent', 'Independent'], ordered=True)
````

````python
# Round all float variables to 2 decimals, for better stardatization (all the real observation would have no more than 2 decimals)

# Step 1: List of columns to be rounded
columns_decimal = ['BMI', 'ADL', 'CholesterolTotal','CholesterolLDL','CholesterolHDL','CholesterolTriglycerides']
columns_integer = ['AlcoholConsumption', 'PhysicalActivity', 'DietQuality', 'SleepQuality', 'MMSE', 'FunctionalAssessment']

# Step 2: Apply rounding to only the selected columns
df[columns_decimal] = df[columns_decimal].round(2)

df[columns_integer] = df[columns_integer].round(0).astype(int)
````

````python
# We will keep thhe dataset for ML anaçysis and create a copy for data wrangling to support dataviz
df_dataviz = df
````

````python
#Many features were converted as integer datatype but most of them are categorical: binary boolean or ordinal
#Explicitly convert the binary categorical variables: first replacing the 'yes' and 'no' values to True and False and then using bool type
#Converting binary object datatypes into binary boolean

# List of columns to convert
columns_to_convert = ['Smoking','FamilyHistoryAlzheimers', 'CardiovascularDisease','Depression','Diabetes','HeadInjury', 'Hypertension', 'MemoryComplaints',
                      'BehavioralProblems', 'Confusion', 'Disorientation', 'PersonalityChanges', 'DifficultyCompletingTasks', 'Forgetfulness', 'Diagnosis']

# Define the conversion mapping
conversion_mapping = {1: True, 2: False}

# Loop over the columns and perform conversion
for column in columns_to_convert:
    df_dataviz[column] = df_dataviz[column].replace(conversion_mapping).astype(bool)
````

````python
# Converting the nominal category Gender
df_dataviz['Gender'] = df_dataviz['Gender'].astype("category")

#Transform the binary feature `gender` into numerical values
df_dataviz['Gender'] = df_dataviz['Gender'].map({
                0:"Female",
                1:"Male"})
````

````python
#Converting "EducationalLevel" feature to ordinal category
educ_level = ["None", "High School", "Bachelor's", "Higher"]

df_dataviz['EducationLevel'] = df_dataviz['EducationLevel'].map({0: "None", 1: "High School", 2: "Bachelor's", 3: "Higher"})

df_dataviz["EducationLevel"] = pd.Categorical(df_dataviz["EducationLevel"], categories=educ_level, ordered=True)
````

````python
#Converting "Ethnicity" feature to nominal category
ethinicity = ["Caucasian", "Black", "Asian", "Other"]

df_dataviz['Ethnicity'] = df_dataviz['Ethnicity'].map({0: "Caucasian", 1: "Black", 2: "Asian", 3: "Other"})

df_dataviz["Ethnicity"] = pd.Categorical(df_dataviz["Ethnicity"], categories=ethinicity, ordered=False)
````

````python
# Create the Dash app
app = dash.Dash(__name__)

# Layout of the app
app.layout = html.Div([
    # Filters
    html.Div([
        html.Label('Select Gender:'),
        dcc.Dropdown(
            id='gender-filter',
            options=[{'label': i, 'value': i} for i in df_dataviz['Gender'].unique()],
            value=None,  # Default is no filter (show all)
            multi=True  # Allow multiple selections
        ),
        
        html.Label('Select Educational Level:'),
        dcc.Dropdown(
            id='escolarity-filter',
            options=[{'label': i, 'value': i} for i in df_dataviz['EducationLevel'].unique()],
            value=None,
            multi=True
        ),
        
        html.Label('Select Ethnicity:'),
        dcc.Dropdown(
            id='ethnicity-filter',
            options=[{'label': i, 'value': i} for i in df_dataviz['Ethnicity'].unique()],
            value=None,
            multi=True
        ),

        html.Label('Select Doctor:'),
        dcc.Dropdown(
            id='doctor-filter',
            options=[{'label': i, 'value': i} for i in df_dataviz['DoctorInCharge'].unique()],
            value=None,
            multi=True
        ),
    ]),

    # Bar chart to show age distribution
    dcc.Graph(id='age-distribution-bar', style={'width': '90%', 'paddingTop': '10px'}),

    # Table to show filtered data, now displayed below the graph
    html.Div(id='table-div', style={'paddingTop': '20px'})
])

# Callback to update table and bar chart based on filters
@app.callback(
    [Output('table-div', 'children'),
     Output('age-distribution-bar', 'figure')],
    [Input('gender-filter', 'value'),
     Input('escolarity-filter', 'value'),
     Input('ethnicity-filter', 'value'),
     Input('doctor-filter', 'value')]
)
def update_table_and_chart(gender, escolarity, ethnicity, doctor):
    # Filter data based on selected values
    filtered_df = df_dataviz.copy()

    if gender:
        filtered_df = filtered_df[filtered_df['Gender'].isin(gender)]
    
    if escolarity:
        filtered_df = filtered_df[filtered_df['EducationLevel'].isin(escolarity)]
    
    if ethnicity:
        filtered_df = filtered_df[filtered_df['Ethnicity'].isin(ethnicity)]

    if doctor:
        filtered_df = filtered_df[filtered_df['DoctorInCharge'].isin(doctor)]

    # Create a table with filtered data
    table = html.Table([
        html.Thead(html.Tr([html.Th(col, style={'color': 'black'}) for col in filtered_df.columns])),
        html.Tbody([
            html.Tr([html.Td(filtered_df.iloc[i][col], style={'color': 'black'}) for col in filtered_df.columns])
            for i in range(min(len(filtered_df), 10))  # Show only first 10 rows
        ])
    ])

    # Create a bar chart for age distribution
    fig = px.histogram(
        filtered_df, 
        x='Age', 
        nbins=10, 
        color='Gender',  # Add color based on gender
        barmode='group',  # Group the bars
        title='Age Distribution by Gender',
        labels={'Age': 'Age of Participants', 'count': 'Number of Participants'},
        color_discrete_sequence=px.colors.qualitative.Set2  # High-contrast color palette
    )

    # Customize layout: add titles, axis labels, and improve appearance
    fig.update_layout(
        xaxis_title="Age",
        yaxis_title="Count",
        legend_title="Gender",
        plot_bgcolor='white',  # Light background for the plot
        paper_bgcolor='#F0F0F0',  # Match the background of the entire page
        font=dict(color='black'),  # Black text for readability
        title_font_size=18,
        font_size=12
    )

    return table, fig

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
````
