## Dataset Description

Neurologisk Klinik has compiled an anonymized dataset [1] with detailed health information for 2,149 patients, each identified by a unique ID (4751 and 6900). The dataset contains various types of data. Patient Information includes a unique PatientID (integer). Demographic Details cover Age (integer), Gender (binary), Ethnicity, and EducationLevel (categorical). Lifestyle Factors include BMI (float), Smoking, AlcoholConsumption, PhysicalActivity, DietQuality, and SleepQuality (integers or binary). Medical History variables are binary (e.g., FamilyHistoryAlzheimers, Hypertension). Clinical Measurements include blood pressure and cholesterol levels (integers). Cognitive and Functional Assessments cover MMSE, FunctionalAssessment, ADL (integers), and several binary indicators. Symptoms are binary (e.g., Confusion, Forgetfulness). Diagnosis Information is binary for Alzheimer’s status. DoctorInCharge: This column contains confidential information about the doctor in charge, with "XXXConfid" as the value for all patients.

https://www.kaggle.com/datasets/rabieelkharoua/alzheimers-disease-dataset

##### Details - Patient Information
1. Patient ID

        1.1 PatientID: A unique identifier assigned to each patient (4751 to 6900).

2. Demographic Details

        2.1 Age: The age of the patients ranges from 60 to 90 years.
        2.2 Gender: Gender of the patients, where 0 represents Male and 1 represents Female.
        2.3 Ethnicity: The ethnicity of the patients, coded as follows:
            0: Caucasian
            1: African American
            2: Asian
            3: Other
        2.4 EducationLevel: The education level of the patients, coded as follows:
            0: None
            1: High School
            2: Bachelor's
            3: Higher
            
3. Lifestyle Factors

        3.1 BMI: Body Mass Index of the patients, ranging from 15 to 40.
        3.2 Smoking: Smoking status, where 0 indicates No and 1 indicates Yes.
        3.4 AlcoholConsumption: Weekly alcohol consumption in units, ranging from 0 to 20.
        3.5 PhysicalActivity: Weekly physical activity in hours, ranging from 0 to 10.
        3.6 DietQuality: Diet quality score, ranging from 0 to 10.
        3.7 SleepQuality: Sleep quality score, ranging from 4 to 10.

4. Medical History

        4.1 FamilyHistoryAlzheimers: Family history of Alzheimer's Disease, where 0 indicates No and 1 indicates Yes.
        4.2 CardiovascularDisease: Presence of cardiovascular disease, where 0 indicates No and 1 indicates Yes.
        4.3 Diabetes: Presence of diabetes, where 0 indicates No and 1 indicates Yes.
        4.4 Depression: Presence of depression, where 0 indicates No and 1 indicates Yes.
        4.5 HeadInjury: History of head injury, where 0 indicates No and 1 indicates Yes.
        4.6 Hypertension: Presence of hypertension, where 0 indicates No and 1 indicates Yes.

5. Clinical Measurements

        5.1 SystolicBP: Systolic blood pressure, ranging from 90 to 180 mmHg.
        5.2 DiastolicBP: Diastolic blood pressure, ranging from 60 to 120 mmHg.
        5.3 CholesterolTotal: Total cholesterol levels, ranging from 150 to 300 mg/dL.
        5.4 CholesterolLDL: Low-density lipoprotein cholesterol levels, ranging from 50 to 200 mg/dL.
        5.5 CholesterolHDL: High-density lipoprotein cholesterol levels, ranging from 20 to 100 mg/dL.
        5.6 CholesterolTriglycerides: Triglycerides levels, ranging from 50 to 400 mg/dL.

6. Cognitive and Functional Assessments

        6.1 MMSE: Mini-Mental State Examination score, ranging from 0 to 30. Lower scores indicate cognitive impairment.
        6.2 FunctionalAssessment: Functional assessment score, ranging from 0 to 10. Lower scores indicate greater impairment.
        6.3 MemoryComplaints: Presence of memory complaints, where 0 indicates No and 1 indicates Yes.
        6.4 BehavioralProblems: Presence of behavioral problems, where 0 indicates No and 1 indicates Yes.
        6.5 ADL: Activities of Daily Living score, ranging from 0 to 10. Lower scores indicate greater impairment.

7. Symptoms

        7.1 Confusion: Presence of confusion, where 0 indicates No and 1 indicates Yes.
        7.2 Disorientation: Presence of disorientation, where 0 indicates No and 1 indicates Yes.
        7.3 PersonalityChanges: Presence of personality changes, where 0 indicates No and 1 indicates Yes.
        7.4 DifficultyCompletingTasks: Presence of difficulty completing tasks, where 0 indicates No and 1 indicates Yes.
        7.5 Forgetfulness: Presence of forgetfulness, where 0 indicates No and 1 indicates Yes.

8. Diagnosis Information

        8.1 Diagnosis: Diagnosis status for Alzheimer's Disease, where 0 indicates No and 1 indicates Yes.

9. Confidential Information

        9.1 DoctorInCharge: This column contains confidential information about the doctor in charge, with "XXXConfid" as the value for all patients.

````python
import pandas as pd
import numpy as np
from skimpy import skim
````

````python
df = pd.read_csv("alzheimers_disease_data.csv", sep=",",true_values =["yes"], false_values=["no"])
````

````python
# PatientID is a category rather than a Integer and will be correctly assigned

df['PatientID'] = df['PatientID'].astype("category")
````

````python
#Many features were converted as 'object' datatype but but most of them are categorical: binary boolean or ordinal
#Explicitly convert the binary categorical variables: first replacing the 'yes' and 'no' values to True and False and then using bool type
#Converting binary object datatypes into binary boolean

# List of columns to convert
columns_to_convert = ['Smoking','FamilyHistoryAlzheimers', 'CardiovascularDisease','Depression','Diabetes','HeadInjury', 'Hypertension', 'MemoryComplaints',
                      'BehavioralProblems', 'Confusion', 'Disorientation', 'PersonalityChanges', 'DifficultyCompletingTasks', 'Forgetfulness', 'Diagnosis']

# Define the conversion mapping
conversion_mapping = {1: True, 2: False}

# Loop over the columns and perform conversion
for column in columns_to_convert:
    df[column] = df[column].replace(conversion_mapping).astype(bool)
````

````python
# Converting the nominal category Gender
df['Gender'] = df['Gender'].astype("category")

#Transform the binary feature `gender` into numerical values
df['Gender'] = df['Gender'].map({
                0:"Female",
                1:"Male"})
````

````python
#Converting "EducationalLevel" feature to ordinal category
educ_level = ["None", "High School", "Bachelor's", "Higher"]

df['EducationLevel'] = df['EducationLevel'].map({0: "None", 1: "High School", 2: "Bachelor's", 3: "Higher"})

df["EducationLevel"] = pd.Categorical(df["EducationLevel"], categories=educ_level, ordered=True)
````

````python
#Converting "Ethnicity" feature to nominal category
ethinicity = ["Caucasian", "Black", "Asian", "Other"]

df['Ethnicity'] = df['Ethnicity'].map({0: "Caucasian", 1: "Black", 2: "Asian", 3: "Other"})

df["Ethnicity"] = pd.Categorical(df["Ethnicity"], categories=ethinicity, ordered=False)
````

````python
df.sample(8)
````

````python
# Set a random seed for reproducibility
np.random.seed(42)

# Include the random Doctor In Charge variable

df['DoctorInCharge'] = np.random.choice(['Med Sven', 'Med Anna', 'Med Erik'], size=len(df))

# Convert the column to categorical
df['DoctorInCharge'] = pd.Categorical(df['DoctorInCharge'], categories=['Med Sven', 'Med Anna', 'Med Erik'])

# Check the distribution
print(df['DoctorInCharge'].value_counts())
````

````python
df.dtypes
````

````python
#Descriptive analysis using skim
skim(df)
````

````python
# Creating variables related to Medications in use for Hyptertension

# Step 1: Create 'HyperMed' column with conditions
df['HyperMed'] = np.where(
    df['Hypertension'], 
    np.random.choice(['A', 'B', 'C'], size=len(df)),  # Randomly assign A, B, or C for 'True'
    'Z'  # Assign 'Z' for 'False'
)

# Step 2: Convert 'HyperMed' to a categorical variable
df['HyperMed'] = pd.Categorical(df['HyperMed'], categories=['A', 'B', 'C', 'Z'])

# Step 3: Function to check if the assignment is correct
def check_hypermed_assignment(df):
    # Check if those with 'Hypertension = True' have only 'A', 'B', or 'C'
    true_check = df[df['Hypertension']]['HyperMed'].isin(['A', 'B', 'C']).all()
    
    # Check if those with 'Hypertension = False' have only 'Z'
    false_check = (df[~df['Hypertension']]['HyperMed'] == 'Z').all()
    
    # Return result
    if true_check and false_check:
        return "Assignments are correct!"
    else:
        return "Assignments are incorrect!"

# Step 4: Run the check
print(check_hypermed_assignment(df))

# Optional: Check distribution of 'HyperMed'
print(df['HyperMed'].value_counts())
````

````python
# Creating variables related to Medications in use for Diabetes

# Step 1: Create 'DiabMed' column with conditions
df['DiabMed'] = np.where(
    df['Diabetes'], 
    np.random.choice(['D', 'E'], size=len(df)),  # Randomly assign D, E for 'True'
    'Z'  # Assign 'Z' for 'False'
)

# Step 2: Convert 'HyperMed' to a categorical variable
df['DiabMed'] = pd.Categorical(df['DiabMed'], categories=['D', 'E', 'Z'])
````

````python
# Deleting features that is not related to the current analysis
columns_to_delete = ["Skin_Cancer", "Other_Cancer", "Arthritis", "Height_(cm)", "Weight_(kg)"]

# Delete the columns and assign a new variable `df_processed`
df_processed = df.drop(columns_to_delete, axis=1)
df_processed.sample(5)
````
